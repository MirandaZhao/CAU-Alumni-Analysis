数据预处理: 

上述数据已经过预处理：数据清理以及合并专业。数据清理使用正则表达式法提取数字和文字，去除无关文字，调整数据格式。同时，为了更能反映不同地区间的专业大类人数比较，采取合并相同大类下细分专业分类的归类方法。合并规则为：将食品科学、食品科学与工程合并为食品科学与工程；将国际金融、银行与金融支持服务、金融学合并为金融学；将信息技术、信息学/信息研究、计算机科学合并为计算机科学；将会计与财务、会计学合并为会计学；将生物科学、生物学合并为生物学。

```python
import re

data = """
所学专业 添加
581 经济学关闭 
337 金融学关闭 
305 食品科学关闭 
285 计算机科学关闭 
247 计算科学关闭 
243 生物科学关闭 
217 动物学关闭 
203 工商管理关闭 
202 统计学关闭 
193 生物学关闭 
190 传播学关闭 
162 兽医学关闭 
154 电气与电子工程关闭 
152 数学关闭 
138 会计及相关服务关闭
"""

# 使用正则表达式提取数字和文字
#正则表达式为 (\d+)\s*([\u4e00-\u9fa5/]+)，其中 (\d+) 匹配数字，\s+ 匹配一个或多个空格， [\u4e00-\u9fa5] 匹配中文字符，([a-zA-Z/]+) 匹配文字
#matches = re.findall(r'(\d+)\s+([a-zA-Z/]+)', data)
#matches = re.findall(r'(\d+)\s*([\u4e00-\u9fa5/]+)', data)
#修改了匹配数字的部分，允许匹配逗号分隔的数字。
matches = re.findall(r'(\d{1,3}(?:,\d{3})*)\s+([\u4e00-\u9fa5a-zA-Z\s&()/-]+)', data)

# 打印结果
for match in matches:
    number, text = match
    text = text.replace("打开", "").replace("开启", "").replace("关闭", "")
    print("{} {}".format(text.strip(), number))  
```

```python
import pandas as pd
from io import StringIO

# 输入数据
data_str = """
专业	专业人数
经济学	581
金融学	337
食品科学	305
计算机科学	285
计算科学	247
生物科学	243
动物学	217
工商管理	203
统计学	202
生物学	193
传播学	190
兽医学	162
电气与电子工程	154
数学	152
会计及相关服务	138
"""

# 将字符串转为DataFrame，去除专业人数中的逗号
#df = pd.read_csv(StringIO(data_str), sep='\t')
df = pd.read_csv(StringIO(data_str.replace(',', '')), sep='\t')

# 将食品科学、食品科学与工程合并为食品科学与工程
df['专业'] = df['专业'].replace({'食品科学': '食品科学与工程'})

# 将国际金融、银行与金融支持服务、金融学合并为金融学
df['专业'] = df['专业'].replace({'国际金融': '金融学', '银行与金融支持服务': '金融学'})

# 将会计与财务、会计学合并为会计学
df['专业'] = df['专业'].replace({'会计与财务': '会计学'})

# 将生物科学、生物学合并为生物学。
df['专业'] = df['专业'].replace({'生物科学': '生物学', '生物技术':'生物学'})

# 将信息技术、信息学/信息研究、计算机科学合并为计算机科学
df['专业'] = df['专业'].replace({'信息学/信息研究':'计算机科学', '计算科学':'计算机科学'})

# 将兽医技术员与兽医助理、兽医学合并为兽医学
df['专业'] = df['专业'].replace({'兽医技术员与兽医助理': '兽医学'})

# 将会计与财务、会计学合并为会计学
df['专业'] = df['专业'].replace({'会计及相关服务': '会计学'})

df = df.groupby('专业', as_index=False).sum()

# 按专业人数从大到小排序
df = df.sort_values(by='专业人数', ascending=False)

# 打印结果
print(df)
```

## 6 结论

6.1 校友人才地区分布趋势

数据显示，全球范围内存在较为广泛的人才流动，国际化程度较高。

中国、美加、欧洲、澳大利亚等地区都吸引了大量校友人才。在海外校友中，美洲校友占比最高。美洲中，选择美东纽约地区和美西加州地区的校友最多。欧洲校友中，选择英国和荷兰的校友最多。

全球就业市场中，中国和美国是热门的就业地区。国内就业市场中，国内的一线城市和较发达省份(北京、上海和广东等)，吸引了大量校友人才，这可能与国家和地区经济规模和发展水平密切相关。

校际教育中，美国普渡大学，荷兰瓦格宁根大学，美国康奈尔大学等是热门的海外学校，这可能与两个学校的优势学科与农大合作较多有关。

全球化使得人才流动更加便捷，各个行业的跨国公司更容易吸引来自不同背景的优秀人才。全球范围内的公司招聘也突显了跨文化工作机会，这对于拥有全球化背景、善于跨文化沟通和协作的校友来说是一项优势。


6.2 校友人才专业分布趋势

不同地区的人才市场呈现出独特的特征，不同地区的校友人才对专业的选择也存在一定同异。

在中国，工商管理、食品科学和计算机领域是最热专业，而海外在经济学、计算机科学、生物学方面较为突出。在海外地区中，美国在计算机科学方面领先，欧洲则更注重金融学和工商管理，澳大利亚是会计和计算机最为热门，日新则是食品和经济金融类专业最受欢迎。而在美国中，美东地区更注重经济学和金融学，而美西地区更集中在计算机科学和生物学领域。此外，技术行业在全球范围内拥有更广泛的分布，生物学和医学领域则在美东地区占据主导地位。侧面反映了不同国家和地区的优势专业、就业市场需求、以及经济和科技发展的地区特色，这为校友提供了更具体的地区和专业选择建议。

全球范围内计算机科学专业的人才需求较高，这与科技行业的蓬勃发展以及数字化转型的推动有关，反映了未来科技发展的方向。出于技术驱动，计算机科学、信息技术等专业吸引了大量校友选择该类专业。该领域人才需求在未来可能会持续增长，也反映了科技发展对各行各业的渗透，深耕相关专业技能可能更有利于职业前景和发展。

可以鼓励校友人才在职业发展中，综合考虑不同地区和学校的优势专业、市场需求、和个人优势及偏好进行申请。注重国际交流和学科深造，提升综合素养和全球化竞争力。

国内文商科类，食品类，动医，计算机类校友，也可多活跃于领英等招聘平台，相关专业人才可能更容易在此类平台中，联结对口校友和招聘机会。


6.3 校友人才领域分布趋势

研究、运营、销售、工程是全球范围内最热门的领域，拥有大量从业人员。这可能与校友所在地区就业资源丰富种类、毕业校友科研水平过硬等因素有关，也可能与学校作为在建世界双一流高校，培养高素质一流人才的定位有关，为教育和科学研究事业输送了大量优质人才。

在计算机科学、信息技术专业人才较多的地区，从事工程、信息技术等领域的比例会更高；在文商科专业人才较多的地区，从事运营、销售等领域的比例会更高。这说明公司更倾向于招聘与其主营业务领域相关的专业人才，这有助于发挥员工在特定领域的专业能力。

美加地区工程和信息技术类领域就业优势较明显相关，同时在该地区高校及科研机构任职的教授人数较多。侧面反映当地劳动力缺口偏好和身份政策倾向。此外，在美国中，美东地区校友更偏好的财务/金融类领域，和当地文商科和金融类专业选择成正相关关系，美西地区更偏好工程和信息技术类领域，与当地校友计算机类专业选择成正相关关系。欧洲地区，运营领域和销售领域人数占比，相较美加地区更加突出，这可能与欧洲地区校友偏好选择会计类，金融类，商业类学科相关。澳洲和日新地区，运营领域和销售领域人数占比，相较美加地区更加突出，与欧洲就业领域结构更相似。

海外校友显示从事教育和研究领域比例更高的原因，可能与不同地区就业机会种类丰富程度不同、海外注册校友停留阶段为在校期间科研经历较多、和海外毕业后选择海外教研领域等原因有关。


6.4 校友人才公司分布趋势

不同行业的公司，在各地区的受欢迎程度有所不同，校友可以根据自身情况结合市场需求，选择适合自己发展的职业路线。

国内校友人才更倾向于在本土知名公司工作，而在美加地区，一些全球性科技巨头更具吸引力，计算机类专业和就业优势较明显。在美国境内，美东地区生物、医学、金融类型就业资源更为丰富有关，美西地区互联网企业就业资源更为丰富。

与农大优势专业对口的企业，如生物医药、农业类、食品快消类等企业，以及有友好合作历史或优势学科对口的知名高校，如中科院、美国普渡、荷兰瓦格宁根大学、哈佛医学院、斯坦福医学院等也成为校友聚集地。这表明高校与高校/公司之间存在紧密的联系，高校培养了对口优势专业的科研和技术人才，为与学校有友好合作历史的学校和企业提供了潜在的招聘渠道。

全球性公司如玛氏食品、先正达种业、雀巢、拜耳等，及互联网公司如亚马逊、谷歌、脸书、字节等公司在全球范围内拥有广泛的员工分布，显示出其在全球市场的强大影响力，促进了全球范围内的人才交流。这些公司在不同地区有不同的语言需求，尤其在欧美和亚洲。因此，具备多语言技能跨国公司和留学背景的求职者更有竞争力。

国内主要的一线城市和一些较发达省份地区，涵盖了全球和国内的知名企业，也成为国内校友就业的主力地区。



校友人才在全球范围内的流动也需要更好地适应不同的文化和工作环境，这对人才本身的综合素养提出了更高的要求。故如果有对所在地区就业的意向，较好的途径之一是申请当地的学校和专业，兼顾就业导向，再在当地择业定居是一条可尝试的职业路径，但这与当地就业市场和身份政策变化也息息相关。

通过全球校友数据分析，我们可以更全面地了解全球范围内不同地区的人才特点、专业偏好、领域趋势、和公司需求。这有助于学校、学生和校友更好地了解专业优势和市场需求，做出更明智的规划和战略决策。
